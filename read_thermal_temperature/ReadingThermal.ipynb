```python
# Compute temperatures of pixels and averaged value for whole object temperature
pixel_temp = fir.extract_thermal_image()
pixel_rows = pixel_temp.sum(axis=0)/len(pixel_temp)
aver_temp = sum(pixel_rows)/len(pixel_rows)
print('image average temperature is:', aver_temp, 'oC')
```
